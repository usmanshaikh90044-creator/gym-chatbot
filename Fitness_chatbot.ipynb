```python
df['Desc'] = df['Desc'].fillna(
    "No detailed description available. Refer to exercise name and type."
)

df['Equipment'] = df['Equipment'].fillna("Not specified")

df['Rating'] = df['Rating'].fillna("Not rated")

df['RatingDesc'] = df['RatingDesc'].fillna("No rating description available")
```

```python
df.isnull().sum()
```

```python
#combining the whole dataset into one column
df['combined_text'] = (
    "Exercise Name: " + df['Title'] + ". " +
    "Description: " + df['Desc'] + ". " +
    "Type: " + df['Type'] + ". " +
    "Body Part: " + df['BodyPart'] + ". " +
    "Equipment: " + df['Equipment'] + ". " +
    "Difficulty Level: " + df['Level'] + ". " +
    "Rating: " + df['Rating'].astype(str) + ". " +
    "Rating Details: " + df['RatingDesc']
)
#cause TF-IDF works on text onot on cols so we combine all the text in one col 
```

```python
import re

def clean_text(text):
    text = text.lower()                      # normalize casing
    text = re.sub(r'\s+', ' ', text)         # remove extra spaces
    text = text.strip()                      # trim edges
    return text

df['combined_text_clean'] = df['combined_text'].apply(clean_text)
```

```python
#vecorization
from sklearn.feature_extraction.text import TfidfVectorizer

```

```python
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    stop_words="english"
)

tfidf_matrix = tfidf.fit_transform(df['combined_text_clean'])
```

```python
tfidf_matrix.shape
```

```python
#connecting the user input to outr data 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#creating a search function
def search_exercises(query, top_k=5):
    # Convert user query to TF-IDF vector
    query_vec = tfidf.transform([query])
    
    # Compute cosine similarity
    similarities = cosine_similarity(query_vec, tfidf_matrix).flatten()
    
    # Get top-k most similar indices
    top_indices = similarities.argsort()[-top_k:][::-1]
    
    # Return matching rows
    return df.iloc[top_indices][
        ['Title', 'BodyPart', 'Equipment', 'Level', 'Desc']
    ]
```

```python
#testing
search_exercises("beginner chest workout at home")
```

```python
#Converting search results into a chat-style response
def format_chat_response(results_df):
    response = "Here are some exercises that match your request:\n\n"
    
    for idx, row in results_df.iterrows():
        response += f"üèãÔ∏è Exercise: {row['Title']}\n"
        response += f"‚Ä¢ Body Part: {row['BodyPart']}\n"
        response += f"‚Ä¢ Equipment: {row['Equipment']}\n"
        response += f"‚Ä¢ Level: {row['Level']}\n"
        response += f"‚Ä¢ Description: {row['Desc']}\n\n"
    
    return response
```

```python
results = search_exercises("beginner chest workout at home")
chat_response = format_chat_response(results)
print(chat_response)
```

```python
import streamlit as st

st.set_page_config(page_title="Gym Chatbot üí™", layout="centered")

st.title("üèãÔ∏è Gym Exercise Chatbot")
st.write("Ask me about workouts, muscles, equipment, or difficulty levels!")

query = st.text_input("Enter your workout query:")

if query:
    results = search_exercises(query)
    chat_response = format_chat_response(results)
    
    st.text(chat_response)
```
